```python
class BST: 
    def __init__(self, value): 
        self.value = value 
        self.left = None 
        self.right = None 
        
    def print_bst(self): 
        array = []
        result = self._depthFirstRepr(array)
        print("-".join(result))
        return 
            
    def _depthFirstRepr(self, array=[]): 
        array.append(str(self.value))
        if self.left: 
            array = self.left._depthFirstRepr(array)
        if self.right: 
            array = self.right._depthFirstRepr(array) 
        return array
            
    
    # time : o(log(n)) average | o(n) worst
    # space : o(log(n)) average | o(n) worst 
    def contains(self, target): 
        if self.value < target: 
            # look on your right 
            if self.right: 
                return self.right.contains(target)
            else: 
                return False
        elif self.value > target: 
            if self.left: 
                return self.left.contains(target) 
            else: 
                return False 
        else: 
            # found the good one 
            return True
        
    # time : o(log(n)) average | o(n) worst
    # space : o(log(n)) average | o(n) worst 
    def insert(self, value): 
        if self.value <= value: 
            if self.right: 
                self.right.insert(value)
            else: 
                # we insert here because we have no children 
                self.right = BST(value)
        else: 
            if self.left: 
                self.left.insert(value)
            else: 
                # same, we have no children on the left and we are looking to insert to the left 
                self.left = BST(value) 
        return self
    
    def remove(self, value, parent=None): 
        
        # we have to consider different use cases : 
        
        # 1 ) we want to remove the root : no parent ! 
            # if left and right = > get min from right sub tree 
            # if only one of them => just get next node 
            # if none, single node => we are done
            
        # 2 ) we want to remove a node that has a left or a write child 
            # get min from right sub tree
            
        # 3 ) We want to remove a node that has no children => tell parent 
        
        
        # we need to find the value 
        print(self.value, value)
        if self.value < value:
            if self.right: 
                self.right.remove(value, parent=self)
            else: 
                return 
        elif self.value > value: 
            if self.left: 
                self.left.remove(value, parent=self) 
            else: 
                return 
        else: 
            # we have found the correct node ! 
            # let's list all use cases 
            print("we have found", value)
            if self.left and self.right: 
                print(f"{value} has left and right !")
                # min from left subtree ! 
                self.value = self.right.getMinValue()
                # we need to remove the node we just replaced ! aka min node 
                print("removing on the right", self.value)
                return self.right.remove(self.value, parent=self)
            # case where the root has only left or right or nothing 
            if not parent: 
                print(f"{value} is a root with one side !")
                if self.left: 
                    # get next node 
                    self.value = self.left.value 
                    self.right = self.left.right 
                    self.left = self.left.left 
                elif self.right: 
                    self.value = self.right.value 
                    self.left = self.right.left 
                    self.right = self.right.right 
                else: 
                    # single node 
                    return 
                    
            # if no left | right children and has parent, it's just a node with either left or right children 
            # and this node is not the root 
            else:
                print(f"{value} is just a node with a parent and one side")
                if parent.left == self: 
                    if self.left: 
                        parent.left = self.left 
                    else: 
                        parent.left = self.right
                elif parent.right == self: 
                    if self.left: 
                        parent.right = self.left 
                    else: 
                        parent.right = self.right
                print("exit")
                return self 
                
    def getMinValue(self): 
        if self.left: 
            return self.left.getMinValue()
        else: 
            return self.value
```

```python
# create a bst 

root = BST(10) 
root.left = BST(5) 
root.left.left = BST(2)
root.left.left.left = BST(1) 
root.left.right = BST(5) 

root.right = BST(15)
root.right.right = BST(22) 
root.right.left = BST(15) 
root.right.left.right = BST(14)

```
